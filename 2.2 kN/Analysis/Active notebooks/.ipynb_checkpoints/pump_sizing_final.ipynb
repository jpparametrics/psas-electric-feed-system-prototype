# Electric Feed System Pump Size Estimates
We're exploring the use of an all-electric pump for the propellant feed system. What follows is a preliminary pump sizing and requirements analysis to guide further inquiry.

## Design Approach
Turbomachines, contrary to popular belief, are laregly not powered by black magic. The process for determining pump requirements is a straight forward one. First we need to establish design goals. High efficiency, high performance, and minimum mass are desirable from a vehicle mass faction standpoint. But designs that increase efficiency and decrease mass can also increase pump-inlet pressure required to suppress cavitation, which can in turn lead to increases in propellant tank mass. Also we do not wish to compromise design simplicity, reliability, operational life, and cost. In some sense these are hard tradeoffs. So we should first define the relative importance/emphasis of these design goals.

## Design Inputs
Engine and vehicle requirements dictate the types, flow rates and pressure levels of the propellants that the pumps deliver to the engine thrust chamber. These are assumed to be given from an earlier and more fundamental analysis of vehicle design. The pumps must deliver propellant without requiring an inlet pressure greater that that allowed by other parts of the feed system. A pump operating under a specified condition (rotational speed and fluid angles of attack on internal surfaces) produce a constant-volume flow rate and constant head rise. To summarize the (given) inputs are propellant mass flow rate, pump pressure rise, pump inlet pressure, and physical and derived quantities such as propellant density and vapor pressure, volume flow rate and NPSH. Volumetric flow rate, Q, is given by

$$
Q = \frac{\dot{m}}{\rho}
$$

where $\dot{m}$ is the mass flow rate, and $\rho$ is the fluid density.

The pump head rise is given by

$$
H_p = \frac{\Delta p_p}{g_0 \rho}
$$

where $\Delta p_p$ is the required rise in pressure and $g_0$ is the gravitational acceleration.

Available Net Positive Suction Head is given by 

$$
NPSH_{available} = \frac{p_i - p_v}{g_0 \rho}
$$

Note that this formulation of $NPSH$ assumes that the elevation head (including the additional head due to acceleration of the launch vehicle) is small relative to the vapor pressure and tank pressure.

Also note that due to the preponderence of barleycorn units used in hydraulic engineering this document will use them as well, both by convention, and also due to the fact that most vendor data sheets are provided in that unit system.

```python
import math as m
import numpy as np
import matplotlib.pyplot as plt

#physical constants
g_0 = 9.81                                           # gravitational acceleration [m/s/s]
R = 8.3145                                           # Universal gas constant [Pa*m^3/K*mol]

# Propellant properties
lox = {'M': 31.999 / 1000,                           # Molar mass [kg/mol]
       'T_e': 90.2,                                  # Propellant temperature [K]
       'C_pl': 1.729 * 1000,                         # Liquid phase specific heat [J/kg*K]
       'C_pv': 0.919 * 1000,                         # Vapor phase specific heat (constant pressure) [J/kg*K]
       'h_v': 213.05 * 1000,                         # Heat of vaporization [J/kg]
       'T_v': 90.2,                                  # Vaporization temperature [K]
       'R_p': R / (31.999 / 1000),                   # Specific gas constant 
       'p_t': (350+70) * 1.1 * 6894,                 # Propellant tank initial pressure [Pa]
       'a': 1.382 / 10,                              # van der Wall constant A [Pa*m^6/mol^2]
       'b': 0.03186 / 1000,                          # van der Wall constant B [m^3/mol]
       'rho': 1141,                                  # Density [kg/m^3]
       'p_v': 101.4E3,                               # Operating point vapor pressure [Pa]
       'name': 'lox'
      }

ipa = {'M': 60.1 / 1000,                             # Molar mass [kg/mol]
       'T_e': 293,                                   # Propellant temperature [K]
       'C_pl': 7.25 * 1000 * (60.1 / 1000),          # Liquid phase specific heat [J/kg*K]
       'C_pv': 1.75 * 1000,                          # Vapor phase specific heat (constant pressure) [J/kg*K]
       'h_v': 36800 * (60.1 / 1000),                 # Heat of vaporization [J/kg] (435.35) 664 kJ/kg @ BP
       'T_v': 355.8,                                 # Vaporization temperature [K]
       'R_p': R / (60.1 / 1000),                     # Specific gas constant 
       'p_t': ((350                                  # Chamber pressure [psi]
             + 145                                   # Regen pressure drop [psi]
             + 70)                                   # Injector pressure drop [psi]
             * 1.1 * 6894),                          # Propellant tank initial pressure [Pa]
       'a': 12.18 / 10,                              # van der Wall constant A [Pa*m^6/mol^2]
       'b': 0.08407 / 1000,                          # van der Wall constant B [m^3/mol]
       'rho': 789,                                   # Density [kg/m^3]
       'p_v': 8.84E3,                                # Operating point vapor pressure [Pa]
       'name': 'ipa'
      }

h20 = {'M': 18.01528 / 1000,                         # Molar mass [kg/mol]
       'T_e': 293.15,                                # Propellant temperature [K] @BP
       'C_pl': 0,                                    # Liquid phase specific heat [J/kg*K]
       'C_pv': 4181.3,                               # Vapor phase specific heat (constant pressure) [J/kg*K]
       'h_v': 104 * 1000,                            # Heat of vaporization [J/kg]
       'T_v': 373.15,                                # Vaporization temperature [K]
       'R_p': R / (18.01528 / 1000),                 # Specific gas constant
       'p_t': (350+70) * 1.1 * 6894,                 # Propellant tank initial pressure [Pa]
       'a': 0.5536,                                  # van der Wall constant A [Pa*m^6/mol^2]
       'b': 0.00003049,                              # van der Wall constant B [m^3/mol]
       'rho': 999.9720,                              # Density [kg/m^3]
       'p_v': 0,                                     # Operating point vapor pressure [Pa]
       'name': 'h20'
      }

# rocket model (assuming sea-level operation)
isp = 246 * 0.90                                     # specific impulse (sea-level, optimistic estimate) [s]
f = 3.2E3                                            # thrust [N]
mdot_t = f/(g_0 * isp)                               # total mass flowrate [kg/s]
chamber_p= 350 * 6.895 * 1000                        # chamber pressure (assumed to be 380 psi currently) [Pa] 
loss_factor = 1.15                                   # estimate of line and injector losses
OF = 1.3                                             # mixture ratio
mdot_o = mdot_t / (1 + (1/OF))                       # oxidizer mass flowrate [kg/s]
mdot_f = mdot_t / (1 + OF)                           # fuel mass flowrate [kg/s]
mdot_h = mdot_t / 2
p_i = 101.3E3*2                                      # inlet pressure (currently 1 atm (gauge)) [Pa]
delta_p = chamber_p * loss_factor - p_i              # required pump discharge pressure [Pa]

# barleycorn conversions
g_0 = 32.2
delta_p *= 0.000145038

def imps(prop):
    rho = prop['rho'] * 0.062428
    h_v = prop['p_v'] * 0.000145038 * 144 / rho
    if prop == ipa:
        mdot = mdot_f * 2.20462
    elif prop == lox:
        mdot = mdot_o * 2.20462
    elif prop == h20:
        mdot = mdot_o * 2.20462
    q = mdot / rho                                   # volumetric flowrate [f^3/s]
    h_p = 144 * delta_p / rho                        # required head rise [ft]
    h_i = p_i * 0.000145038 * 144 / rho              # inlet head [ft]
    npsh_a = h_i - h_v                               # Net Positive Suction Head (available) [ft]
    return rho, h_v, mdot, q, h_p, npsh_a

print("mass flow rate")
print("%.3f" % imps(ipa)[2], "lbm/s (IPA)")
print("%.3f" % imps(lox)[2], "lbm/s (LOX)", "\n")
print("volumetric flow rate")
print("%.4f" % imps(ipa)[3], "ft^3/s (IPA)")
print("%.4f" % imps(lox)[3], "ft^3/s (LOX)", "\n")
print("required pressure head")
print("%.3f" % imps(ipa)[4], "ft (IPA)")
print("%.3f" % imps(lox)[4], "ft (LOX)", "\n")
print("Net Positive Suction Head Available")
print("%.3f" % imps(ipa)[5], "ft (IPA)")
print("%.3f" % imps(lox)[5], "ft (LOX)")
```

## Design Outputs
Given the requirements and the design inputs there are six basic steps needed to size a pump.

1. Determine the number of stages.
2. Determine the pump rotational speed.
3. Determine the pump impeller tip speeds.
4. Determine the pump impeller entrance and exit tip diameters
5. Determine pump efficiency
6. Determine the shaft power required to drive the pump.


## Determine Number of Stages
The number of stages is generally either determined by the next largest integer value greater than the ratio of the pump-pressure rise to the maximum allowable stage-pressure rise, or by $NPSH_{required}$. If a centrifugal pump is used the approximate limits on stage-pressure rise is  47MPa for LOX and ethanol (citation needed). At 47 MPa impeller tips become so thick that they restrict flow passages and develop large k-losses. We determine the number of stages from a simple rule of thumb

$$ \label{eqn:1}
n \geq \frac{\Delta p_p}{\Delta p_s}
$$

where $\Delta p_p$ is the required pressure rise and we assume $\Delta p_s$ to be 47MPa.

```python
# number of stages

delta_p_s = 47.0E6 * 0.000145038       # estimated allowable pressure rise per stage [Pa]
n = int(m.ceil(delta_p/delta_p_s))     # number of stages

print("number of stages")
print(n)
```

## Determine Pump Rotational Speed
The next step is to estimate the impeller-tip speeds, rotational speed, and finally impeller sizes. Impeller tip speed is given by

$$
u_t = \psi \sqrt{\frac{2 g_0 H_p}{n}}
$$

where $\psi$ is the pump-stage head coefficient which we can assume to have a numerical value of between 0.4 and 1.5 (citation needed). As mentioned previously we want the highest rotational speed to get the lowest pump mass and highest performance, however there are practical constraints on things such as bearings. If there is no boost stage and and if there is a definite NPSH limit the pump rotational speed in $rad/s$ can be determined by 

$$
N_r = \frac{u_{ss}NPSH^{0.75}}{\sqrt{Q}}
$$

Suction-specific speed, $u_{ss}$, indicates the minimum net positive suction head (NPSH, inlet pressurization above the vapor pressure expressed in meters) at which a pump operating at rotational speed $N$, and volume flow rate $Q$, can operate without cavitation effects. $u_{ss}$ values fall between 5000 and 50,000 depending on efficiency of the pump with 7000 to 12,000 being typical. Lower numbers are more conservative. A large suction-specific speed indicates an ability to operate at low inlet pressures. Large values are obtained by using large diameters for the pump-inlet tip and small diameters for the inlet hub such that the inlet axial velocity head is minimized. Additionally thin, gradually curved blades with sharp leading edges produce the minimum static pressure gradients on the blade surfaces. Inducers have both features, and are therefore widely used in pumping applications on rocket engines. The important thing is that the $NPSH_{available}$ must be higher than $NPSH_{required}$ to avoid cavitation. $NPSH_{required}$ is determined on a pump by pump basis. This places the upper limit on impeller specific speed. If additional head is required by the pump  the propellant needs to be pre-pressurized before reaching the pump impeller, such as with the aformentioned inducer, or with additional pump stages (e.g. 'booster' stages), or by pressurizing the propellant tank. As a historical note the German V-2 oxygen tank was pressurized to 2.3 atm (autogenuously) largely to suppress pump cavitation.

$$
N_s = \frac{N_r Q^{\frac{1}{2}}}{\left(\frac{H_p}{n}\right)^{\frac{3}{4}}}
$$

$N_s$ is the stage-specific speed where $N_r$ is the pump's rotational speed and $n$ is the number of stages. Stage specific speed, $N_s$, is a parameter that characterizes geometrically and hydrodynamically similar pumps. Centrifugal pumps are generally low-capacity/high-head devices and thus are at the low specific speed side of the spectrum. Each class of pump (radial, francis, mixed, axial) lends itself to a particular range of specific speeds.


```python
#figure from Munson
from IPython.display import Image
Image(filename='specific_speed.PNG')
```

```python
# pump rotational speed

def pump_speed(prop):
    psi = 1                                                            # pump stage head coefficient (estimated)
    u_t = psi * m.sqrt(2 * g_0 * imps(prop)[4])               # impeller vane tip speed [ft/s]
    u_ss = 7000                                                        # suction specific speed
    npsh_r = imps(prop)[5] * 0.5
    n_r = (u_ss * npsh_r**0.75)/(21.2 * m.sqrt(imps(prop)[3]))  # pump rotational speed [RPM]
    n_r_rad = n_r * 2 * m.pi / 60
    n_s = (21.2 * n_r * m.sqrt(imps(prop)[3]))/(imps(prop)[4] / n)**0.75   # pump specific speed
    return u_t, n_r, n_s, n_r_rad

print("impeller tip speed")
print("%.2f" % pump_speed(ipa)[0], "ft/s (IPA)")
print("%.2f" % pump_speed(lox)[0], "ft/s (LOX)", "\n")
print("pump rotational speed")
print("%.2f" % pump_speed(ipa)[1], "RPM (IPA)")
print("%.2f" % pump_speed(lox)[1], "RPM (LOX)", "\n")
print("pump specific speed")
print("%.2f" % pump_speed(ipa)[2], "(IPA)")
print("%.2f" % pump_speed(lox)[2], "(LOX)")
```

It should be noted that the combination of flowrate, pressure rise and fluid density togeather with the allowable range of speeds places the pump in the very low end of the specific speed range, where the predicted maximum attainable efficicency for several types of centrifugal pumps are below those of positive displacement type pumps. It is not likely that an efficiency beter than 50% is attainable; as such several stages may be required for the desired capacity. Positive dispalcement type pumps may offer improved efficiency at the cost of less steady flow, noise, and much poorer reliability.

Additionally, it's difficult to extrapolate pump efficiencies availabile in literature to such small high-head pumps, as the implicit assumption of geometric and hydrodynamic similarity may well be broken.

## Impeller Diameters
Knowing the rotational speed, we can get the  impeller diameters using

$$ \label{eqn:5}
D_o = \frac{u_t}{N_r}
$$
$$ \label{eqn:6}
D_i = \left( \frac{4Q}{\pi \phi N_r(1 - L^2)} \right)^{1/3}
$$

Where $D_o$ is the diameter of the impeller outlet ($m$), also called the discharge, $D_i$ is the diameter of the impeller inlet ($m$), $L$ is the ratio of the hub diameter to the tip diameter, and $\phi$ is the inducer-inlet flow coefficient. Minimizing the factors $\phi$ and $L$ minimizes the NPSH required by the pump. The inducer-inlet flow coefficient, $\phi$, is the tangent of the flow angle approaching the inducers blade tip. It is one of the main design parameters used to maximize suction performance. Generally, lower values are better. Assume 0.1 if using inducer, otherwise use 1.

```python
# impeller diameters

def diameters(prop):
    l = 0.3                                                        # hub to tip diameter ratio (assume 0.3)
    phi = 1                                                        # inducer flow coefficicent (assume 0.1 with inducer, 1.0 otherwise)
    d_o = pump_speed(prop)[0] * 2 / pump_speed(prop)[3] * 12       # impeller outer diameter [in]
    d_i = (4 * imps(prop)[3]/(m.pi * phi * pump_speed(prop)[3] * (1 - l**2)))**(1/3) * 12  # impeller inner diameter [in]
    return d_o, d_i

print("impeller discharge diameter")
print("%.3f" % diameters(ipa)[0], "in (IPA)")
print("%.3f" % diameters(lox)[0], "in (LOX)", "\n")
print("impeller inlet diameter")
print("%.4f" % diameters(ipa)[1], "in (IPA)")
print("%.4f" % diameters(lox)[1], "in (LOX)")
```

## Shaft Power
The required shaft power can be determined from

$$ \label{eqn:7}
P_{req} = \frac{g_0 \dot{m} H_p}{\eta_p}
$$

where $\eta_p$ is the pump efficiency and can be determined from the stage-specific speed and tabulated efficiency data.

```python
# shaft power

from numpy import interp

#from a table of efficiencies for radial impellers found in Munson et. al., 2009 
def effic(n_s):
    eta_t = [0.55, 0.80, 0.83, 0.85, 0.84, 0.82]
    n_s_t = [500, 1000, 1500, 2000, 2500, 3000]
    eta = interp(n_s, n_s_t, eta_t)
    return eta

eff = 0.4 # a more realistic estimated efficiency
def power(mdot, h_p, n_s):  # required shaft power [W]
    p_req = mdot * h_p/(effic(n_s) * 0.738)
    return p_req

def torque(prop):
    T = 30*power(imps(prop)[2], imps(prop)[4], pump_speed(prop)[2])/(m.pi*pump_speed(prop)[1])
    return T

print("pump efficiency")
print("%.2f" % float(100*effic(pump_speed(ipa)[2])), "% (IPA)")
print("%.2f" % float(100*effic(pump_speed(lox)[2])), "% (LOX)", "\n")
print("required shaft power (optimistic)")
print("%.2f" % power(imps(ipa)[2], imps(ipa)[4], pump_speed(ipa)[2]), "W (IPA)")
print("%.2f" % power(imps(lox)[2], imps(lox)[4], pump_speed(lox)[2]), "W (LOX)", "\n")
print("required shaft power (realistic)")
print("%.2f" % power(imps(ipa)[2], imps(ipa)[4], eff), "W (IPA)")
print("%.2f" % power(imps(lox)[2], imps(lox)[4], eff), "W (LOX)", "\n")
print("full load torque")
print("%.2f" % torque(ipa), "kN*m (IPA)")
print("%.2f" % torque(lox), "kN*m (LOX)")
```

It should be noted that 60% is very optimistic for any hombrew pump volute/inducer/impeller combination. We should count ourselves lucky if we hit 40% efficient.

```python
f_t = 50                                              #engine run time [s]
print("Total stored energy")
print("%.2f" % float(power(imps(ipa)[2], imps(ipa)[4], eff)*f_t/1000), "kJ (IPA)")  #Total stored energy [kJ]
print("%.2f" % float(power(imps(lox)[2], imps(lox)[4], eff)*f_t/1000), "kJ (LOX)")  #Total stored energy [kJ]
#changed for illustrative purposes
```

## History:

The Barske Impeller is a design developed by [] Barske in Germany prior to WWII specifically for the high head rise/ low flow rate applications common to the aerospace industry. Barske continued his work in Britain after the war and published his designs in 1955 and 1960.

"To a skilled designer the pump which forms the subject of this paper will, at first glance, appear most unfavorable and may well be regarded as an offense against present views of hydrodynamics"  - Barske of his own designs

## Design:

The Barske impeller is an open faced partial emmision (PE) design which is housed in a concentric volute typically with a single emmision throat with a conical diffuser section oriented tangentally to the case bore. A typical Barske impelelr design is illustrated in the above figure taken from Lobanof & Ross. Other notable characteristics of the Barske impeller are the exceptionally tall impeller tip height ($b_2$), and the fact that the blade angle ($\Theta$) is relatively unimportant except to moderate the flow area of the impeller.

## Inputs:

$\psi \equiv$ pump head coefficient  
   Typically 0.7 to 0.75 for this type of pump.
   
$\phi \equiv$ Pump discharge coefficient.  
   It is possible for this to be $\phi > 1$ which means that the discharce velocity is outrunning the impeller tip speed. this is an artifact of converting the bowl static head into velocity head. 
   
$N \equiv$ pump rotational speed
   As defined by the motor selection and flowrate requirements
 
$\dot{m} \equiv$ fuel and oxydizer mass flowrates as defined by the engine requirements and the fuel mixture ratio

$N_s \equiv$ pump specific speed
   A dimensionless parameter related to efficiency and $d_1$/$D_2$. Typically 400-1000 for PE pumps, peak efficiency around 800
   
$S_s \equiv$ suction specific speed
   Another dimensionless parameter relating to $H$/$NPSH_r$. Typically 7500-10000 for PE pumps

## Outputs:

$D_1 \equiv$ impeller inlet diameter  
   Typically slightly larger than the eye diameter

$D_e \equiv$ impeller eye diameter

$D_2 \equiv$ impeller overall diameter

$d_1 \equiv$ emmision throat diameter.

$b_2 \equiv$ impeller tip height  
   Typically set to at least or moderately greater than the emmision throat diameter ($d_1$).  
   
$\theta \equiv$ blade angle   
   Not super important as long as the flow area remains constant.
   
## Advantages:

* This type of pump does not require close tolerances with the pump casing to provide reasonable performance which will aid both in avoiding rubbing and therefore explosions, as well as ease of manufacuring. 
* The open design reduces the effect of performance decay thus increasing the usable life of the pump which is beneficial for our shoestring budget.
* Designed for High head rise - Low flowrate conditions.
* Can be manufactured with in house tools, tolerances to limit excessive losses are easily achieved.




## Diffuser angle:

Conical diffusers offer a high recovery efficiency because of their minimal wetted angle. A cone angle ($\alpha$) of $10^o$ is common.

## PE impeller calculations:

Theoretical head rise (H') is given as the following:
$$
H' = \frac{u_2^2 - u_1^2}{2g} + \frac{u_2^2}{2g}
$$

Experimental results of static pressure measurements determine that this equation reduces to the following:

$$
H' = \frac{u^2}{2g}+\frac{u^2}{2g} = \frac{u^2}{g}
$$

Where $u$ is taken to be the impeller tip speed. The actual head rise becomes:

$$
H = \psi \frac{u^2}{g}
$$

**NOTE:** recall that $\psi$ is typically in the range of 0.7 to 0.75 for this type of impeller.

Through the magic of some mixture theory that I don't fully understand, but would like to... Making the assumptions that $\eta_d = 0.8$ and $\phi = 0.8$ the above equation can also be written as:

$$
H = \frac{u^2}{2g} + \eta_d \frac{(\phi u)^2}{2g} [1-(\frac{A_1}{A_2})^2] = 0.74 \frac{u^2}{g}
$$

where $\frac{A_1}{A_2}$ is the diffuser area ratio. this means that $\psi = 0.74$ which is solidly in the expected range. This also means that the majority of the developed static pressure head is developed by the forced vortex.

Impeller tip speed is defined as:

$$
u_2 = \pi D_2 N
$$

Subsequently the pump specific speed is defined from the flow rate ($Q$), Head rise ($H$), and input RPM.

$$
N_s = \frac{NQ^{0.5}}{H^{0.75}}
$$

**NOTE:** scalar changes in the flow rate and head rise result in no change in the specific speed.

### Some substitutions:

Substituting $u$ into the equation for head rise ($H$):

$$
H = \psi \frac{(\pi D_2 N)^2}{g}
$$

Similarly for the output flowrate ($Q$):

$$
Q = \phi (\pi D_2 N) A_1 = \phi \pi^2 D_2 N d_1^2
$$

This means that the specific speed($N_s$) then becomes:

$$
N_s = \frac{\phi^{0.5} g^{0.75} d_1}{\psi^{0.75} \pi^{0.5} D_2}
$$

**NOTE:** that the specific speed is independent of rotational speed and is instead defined by the ratio of emmission throat diameter to impeller diameter.  
Given prescribed parameters for $H$, $Q$, and $N_s$ equations determining $d_1$, and $D_2$ can be found.

$$
d_1 = (\frac{Q}{\phi \pi})^{0.5}(\frac{\psi}{H g})^{0.25}
$$

$$
D_2 = (\frac{1}{\pi N})(\frac{H g}{\psi})^{0.5}
$$

```python
# Some design chriteria from what is likely possible:
N = 30000                                             # Predicted RPM deliverable by the motor at given power requirements
psi = 0.7                                              # Pump head coefficient. Standard starting point for Barske impellers
phi = 0.8                                              # Pump discharge coefficient. 
g_0 = 32.2

# Pump dimensions: CURRENTLY MORE CORRECT!!!
def barske_dim(prop, N):
 N_rad = N*2*m.pi/60                                                                               # Convert N to rad/s
 N_s = 21.2*N*(m.sqrt(imps(prop)[3]))/((imps(prop)[4])**(0.75))                                    # Pump specific speed
 d_D = N_s/5500                                                                                    # diameter ratio (d_1/D_2)
 d_1 = ((((imps(prop)[3]*448.831)/(phi*m.pi))**(0.5))*(((psi)/(imps(prop)[4]*g_0))**(0.25)))       # discharge throat diameter
 D_2 = 12*(3600*(imps(prop)[4]*g_0)/(psi*(N*m.pi)**2))**(0.5)                                      # Impeller diameter
 d_1 = D_2*d_D                                                                                     # discharge throat diam.(alt.)
 
 return d_1, D_2, N_s, d_D, N_rad
print("Discharge throat diameter:")
print("%.4f" % barske_dim(ipa, N)[0], " [in] (IPA)")
print("%.4f" % barske_dim(lox, N)[0], " [in] (LOX)", "\n")
print("Impeller diameter:")
print("%.4f" % barske_dim(ipa, N)[1], " [in] (IPA)")
print("%.4f" % barske_dim(h20, N)[1], " [in] (WATER)")
print("%.4f" % barske_dim(lox, N)[1], " [in] (LOX)", "\n")
print("Specific speed:")
print("%.f" % barske_dim(ipa, N)[2], " (IPA)")
print("%.f" % barske_dim(lox, N)[2], " (LOX)", "\n")
print("diameter ratio(d_1/D_2)")
print("%.4f" % barske_dim(lox, N)[3])
```

### Inlet eye diameter:

Optimum eye diameter sizing has been shown to improve efficiency by up to $4 \%$ over larger inlet eye diameters. the trade off comes in the form of lower $NPSH_r$ requirements for impellers with a larger inlet eye diameter. the equation below shows the $NPSH_r$ as a function of inlet eye diameter note the fourth power relationship. For an inducerless design with plenty of $NPSH_a$ eye diameter should be as near optimal as possible

$$
NPSH_r = [\frac{N Q^{0.5}}{S_s}]^{1.333} + \frac{Q^2}{386D_e^4}
$$
  
By analytical means the optimal inlet eye diameter has been determined to be:

$$
D_e = 5.1(\frac{Q}{N})^{0.333}
$$

I will continue to use the assumption that this is valid until a better equation comes along.



```python
# code section for the inlet eye diameter goes here.
def eye(prop):
 D_e = 5.1*(448.831*imps(prop)[3]/N)**(1/3)                  # Optimal inlet eye diam. convert Q from [ft^3/s] to [GPM]

 return D_e

print("Optimal inlet eye diameter:")
print("%.3f" % eye(ipa), " [in] (IPA)")
print("%.3f" % eye(lox), " [in] (LOX)")
```

### Suction specific speed:
Suction specific speed for PE type pumos is expected to be in the range of 7500-10000. Higher suction specific speed means better suction performance, this directly translated to low $NPSH_r$. High $S_s$ values are typically associated with low flow rates, this is a good omen for our particular pump design.

$$
S_s = \frac{NQ^{0.5}}{NPSH_r^{0.75}}
$$

We can leave the above equation in its present form as both $NPSH_r$, and $Q$ are determined by the output requirements and $N$ will be an optimization of the drive motor in compination with the requirements determined above.

### Thoma parameter:
The Thoma parameter, denoted as either $\sigma$ or $\tau$ depending on the text, relates $NPSH_r$ to $H$

$$
\sigma = \frac{NPSH_r}{H} = (\frac{N_s}{S_s})^{1.333}
$$

## Required power:

Similar to the above method the required power for the pump is calculated by the following equation:

$$
P_{req} = \frac{g_0 \dot{m} H_p}{\eta_p}
$$

where $\eta_p$ is the pump efficiency. for now we will take the conservative assumtion that pump efficiency will be around $40  \%$


```python
# Required power:

eta_p = 0.40                                 # assumed pump efficiency 40% common for barske impellers but may be as high as 55%

def power_barske(prop):
    fluid_power = imps(prop)[2]*imps(prop)[4]/737.56
    shaft_power = fluid_power/eta_p
    return fluid_power, shaft_power

print("Fluid power:")
print("%.4f" % power_barske(ipa)[0], "[kW] (IPA)")
print("%.4f" % power_barske(lox)[0], "[kW] (LOX)", "\n")
print("Shaft power:")
print("%.4f" % power_barske(ipa)[1], "[kW] (IPA)")
print("%.4f" % power_barske(lox)[1], "[kW] (LOX)", "\n")
```

## NOTES:  
  
### A note on inducers:

Inducers are high specific speed axial flow stage. Typically with $N_s = 4000-9000$. The inducer must generate enough head to satisfy the needs of the radial stage which established a new lower $NPSH_r$ for the radial stage. Inducer performance is typically taken as the suction specific speed that corresponds to $3 \%$ of the pump head depression.  
  
The current expectation is that an inducer is unnecessary for our current pump design. if it becomes apparent that this assumption is wrong it will significantly increase the manufacuring dificulty involved in designing our impeller.

### A note about non-concentric bowl designs:

For $N_s =$ 800-1000 exchanging the concentric bowl design for a volute collector design can yeild about $6 \%$ advantage but this advantage fades to pairity with the concentric design for $N_s$ in the 300-400 range. The other advantage of a non-concentric bowl design is a significant reduction in radial side loads however this really only effects our bearing selection.

# Mechanical design:

## Vane number:

According to a majority of literature the appropriate number of vanes is somewhat arbitrary depending on the system configuration. 
There are several important things to consider; Shaft operating speed, inlet to outlet area ratio, vane material and strength, pumped fluid force.  

**Shaft operating speed:** It will be important to consider the pump's desired operating speed in order to avoid introducing combustion instabilities by having too few vanes or by setting up sympathetic vibrations in in relation to the plumbing.  

**Inlet to outlet area ratio:** Area ratios must remain the same throughout the impeller.  

**Vane material and strength:** Material strength will give the minimum vane thickness and address geometry concerns.

## Shaft critical speeds:

One critical path problem that will need to be solved is an estimation of the first critical shaft speed. We will need to make sure that operating RPM is lower than the first critical speed in order to avoid introducing resonant vibration or combustion instability into the system durring start up and opertion.

Using Raleigh's equation shown below to estimate the first shaft critical speed:

$$
\omega_1 = \sqrt{\frac{g \sum{w_i y_i}}{\sum{w_i y_i^2}}}\tag{Shigley's Eq. 7.23}
$$

Where:  
$w_i \equiv$ the weight of the $i^{th}$ section of the shaft.  
$y_i \equiv$ the statis deflection of the $i^{th}$ section of the shaft.  

Raleigh's equation however tends to overestimate the first critical speed therefore numbers should be verified using Dunkirley's equation below: 

$$
\frac{1}{\omega_1^2} \approx \sum_{i=1}^{n}{\frac{1}{\omega_{ii}^2}}\tag{Shigley's Eq. 7.32}
$$

```python
# Outputs
def outputs(prop):
    global RPM, pump_efficiency, tip_diameter, head_coefficient, flow_coefficient, suction_specific_speed, diffuser_exit_ratio, diffuser_half_angle, \
           tip_height, blade_number, blade_thickness

    specific_speed = RPM * (imps(prop)[3] * 448.813) ** 0.5 / imps(prop)[4] ** 0.75

    fluid_power = (imps(prop)[2] * 0.453) * (imps(prop)[4] / 3.280) * 9.81 / 1000                        # [kW]

    motor_power_required = fluid_power / pump_efficiency                                                 # [kW]

    specific_diameter = (tip_diameter * imps(prop)[4] ** 0.25) / (imps(prop)[3] * 448.813) ** 0.5

    impeller_diameter = (1300 / (specific_speed * (imps(prop)[4] ** 0.25))) * ((imps(prop)[3] * 448.813) / head_coefficient) ** 0.5

    emission_throat_diameter = 0.268 * m.sqrt((imps(prop)[3] * 448.813) / flow_coefficient) * (head_coefficient / imps(prop)[4]) ** 0.25

    impeller_tip_speed = (m.pi * impeller_diameter * RPM) / 720

    eye_diameter = 5.1 * ((imps(prop)[3] * 448.813) / RPM) ** 0.333

    impeller_eye_area = m.pi * eye_diameter ** 2

    inlet_blade_tip_height = impeller_eye_area / (m.pi * eye_diameter - blade_number * (blade_thickness * 0.03937))

    npshr = RPM * (imps(prop)[3] * 448.813) ** 0.5 / suction_specific_speed ** 1.333 + (imps(prop)[3] * 448.813) ** 2 / (386 * eye_diameter ** 4)

    diffuser_exit_diameter = 2 * m.sqrt(diffuser_exit_ratio) * emission_throat_diameter / 2

    diffuser_length = (diffuser_exit_diameter / 2 - emission_throat_diameter / 2) / m.tan(diffuser_half_angle * m.pi / 180)

    return specific_speed, fluid_power, motor_power_required, specific_diameter, \
        impeller_diameter, emission_throat_diameter, eye_diameter, npshr, diffuser_exit_diameter, \
        diffuser_length, impeller_eye_area, inlet_blade_tip_height, impeller_tip_speed,
```
